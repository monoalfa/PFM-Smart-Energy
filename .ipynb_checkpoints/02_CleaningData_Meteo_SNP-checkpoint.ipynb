**3. ¿Son todas las columnas relevantes?** 

```python
# Eliminamos las filas sobrantes
dataset = dataset.drop(dataset.index[:142])
dataset = dataset.drop(dataset.index[-137:])

# Eliminamos las columnas que no aportan ninguna información extra
dataset = dataset.drop(['PeriodStart', 'Period'],axis=1) #,'AlbedoDaily', 'SnowDepth','SurfacePressure'
dataset
```

**4. ¿El formato de los datos es correcto?** 

```python
# Transformo time_series a formato de fecha y hora con .to_datetime()
dataset['PeriodEnd'] = pd.to_datetime(dataset['PeriodEnd'], format='%Y-%m-%dT%H:%M:%S%z')
# Elimino la zona horaria
dataset['PeriodEnd'] = dataset['PeriodEnd'].dt.tz_localize(None)
# Convierto en el index a la fecha y hora 'time_series'
dataset = dataset.set_index('PeriodEnd')

# Transformo todos los atributos a variables continuas
dataset=dataset.astype('float')

dataset.head()
```

```python
dataset.dtypes
```

Todos los atributos son variables numéricas contínuas (float64). Esto se va a tener en cuenta de cara a la elección y creación del modelo de aprendizaje.
