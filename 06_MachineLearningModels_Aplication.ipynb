## **4. Dataset modification**

En este apartado, se modifica el dataset original con la finalidad de definir un escenario futuro donde hay un incremento de las energías renovables en la isla de Tenerife. El objetivo es analizar un escenario futuro y ver que impacto económico supone la integración o aumento de las renovables en el sistema eléctrico canario.

El modelo ML que hemos desarrollado es una herramienta útil de planificación de la red eléctrica y puede ser clave para la justificación de inversión de plantas renovables. 



```python
x_EO=50/100;
x_FV=5/100;
x_Demand=0.8/100;

Array_EO=df_ML['Generacion_EO_MW']*x_EO
Array_FV=df_ML['Generacion_FV_MW']*x_FV
Array_Demand=df_ML['Demanda_programada_MW']*x_Demand

df_SC=pd.DataFrame(index=df_ML.index)
#df_SC['Generacion_no_renovable_MW']=df_ML['Generacion_no_renovable_MW']-(Array_EO+Array_FV)+Array_Demand
df_SC['Generacion_renovable_MW']=df_ML['Generacion_renovable_MW']+(Array_EO+Array_FV)
df_SC['Demanda_real_MW']=df_ML['Demanda_real_MW']+Array_Demand
df_SC['Generacion_FV_MW']=df_ML['Generacion_FV_MW']+Array_FV
df_SC['Generacion_EO_MW']=df_ML['Generacion_EO_MW']+Array_EO
df_SC['Generacion_TG_MW']=df_ML['Generacion_TG_MW']
df_SC['Demanda_programada_MW']=df_ML['Demanda_programada_MW']+Array_Demand
df_SC['Aportacion_renovable_%']=df_SC['Generacion_renovable_MW']/df_SC['Demanda_real_MW']*100
df_SC['Month']=df_ML['Month']
df_SC['Hour']=df_ML['Hour']
df_SC['Season_i']=df_ML['Season_i']
df_SC=df_SC.drop(['Demanda_real_MW'],axis=1)
df_SC.sample(10)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.neural_network import MLPRegressor
import xgboost as xgb
from sklearn import neighbors
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

Construimos un vector con los valores estimados por el modelo KNR cuando se introducen los datos reales del 2019 y otro con los valores estimados cuando se introducen los datos del escenario definido para el 2021.

Estos vectores serán de utilidad para calcular el coste de generación mensual de la red eléctrica de Tenerife en el escenario actual frente al escenario definido de aumento de renovables.

```python
# Input data for the model ML
X1 = df_ML_f3[best_features3]
X2 = df_SC[best_features3]

# Normalization of train and test data using mean and variance of the training:
X1 = scalerX.transform(X1)
X2 = scalerX.transform(X2)

# Saving the outputs data from the ML
y19_KNR = scalery.inverse_transform(KNR_Model.predict(X1))
y21_KNR = scalery.inverse_transform(KNR_Model.predict(X2))
```

Construimos un nuevo dataset con la información de interés y los costes de generación estimados por nuestro modelo KNR

```python
df_results=pd.DataFrame(index=df_ML.index)

df_results['Demanda_2019_GWh'] = df_ML['Demanda_real_MW']
df_results['Precio_real_2019_€MWh'] = df_ML['Precio_generacion_€MWh']
df_results['Precio_KNR_2019_€MWh'] = y19_KNR
df_results['Demanda_2022_GWh'] = df_ML['Demanda_real_MW']+Array_Demand
df_results['Precio_KNR_2022_€MWh'] = y21_KNR
df_results_monthly=df_results.resample('1M').mean()

```

```python
df_results=df_results.resample('1H').mean()
df_results['Coste_real_2019_M€'] = df_results['Demanda_2019_GWh']*df_results['Precio_real_2019_€MWh']/1000000
df_results['Coste_KNR_2019_M€'] = df_results['Demanda_2019_GWh']*df_results['Precio_KNR_2019_€MWh']/1000000
df_results['Coste_KNR_2022_M€'] = df_results['Demanda_2022_GWh']*df_results['Precio_KNR_2022_€MWh']/1000000
df_results = df_results.resample('1M').sum()
df_results = df_results.drop(['Precio_real_2019_€MWh','Precio_KNR_2019_€MWh','Precio_KNR_2022_€MWh'],axis=1)
df_results['Demanda_2019_GWh'] = df_results['Demanda_2019_GWh']/1000
df_results['Demanda_2022_GWh'] = df_results['Demanda_2022_GWh']/1000
df_results.loc[12]=[df_results['Demanda_2019_GWh'].sum(),df_results['Demanda_2022_GWh'].sum(),df_results['Coste_real_2019_M€'].sum(),df_results['Coste_KNR_2019_M€'].sum(),df_results['Coste_KNR_2022_M€'].sum()]
month = pd.Series(['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Setiembre','Octubre','Noviembre','Diciembre','TOTAL'])
df_results.set_index(month)
```
